```python
from itertools import islice
def take (n,iterable):
    "Return first n items of the iterable as list"
    return list(islice(iterable,n))
```

```python
# converting to dic format and 20 words

affinity_scores = affin.set_index('word')['value'].to_dict()
take(20,affinity_scores.items())
```

```python
nlp = spacy.load("en_core_web_sm")
sentiment_lexicon = affinity_scores

def calculate_sentiment(text:str = None):
    sent_score =0
    if text:
        clean = nlp(text)
        for word in clean:
            sent_score += sentiment_lexicon.get(word.lemma_,0)
    return sent_score
```

```python
# calculating the sentiment values for each sentence

#dataframe['sentiment_value'] = dataframe['clean'].apply(calculate_sentiment)
dataframe['sentiment_value'] = dataframe['clean'].apply(calculate_sentiment)
```

```python
# calculating the number of words in each sentance 

dataframe['word_count'] = dataframe['clean'].str.split().apply(len)
dataframe['word_count'].head(25)
```

```python
# sort based on sentiments value 

dataframe.sort_values(by = 'sentiment_value').head(25)
```
