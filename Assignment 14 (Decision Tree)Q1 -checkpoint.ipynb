```python
# IMPORTING THE LIBRARIES 

import pandas as pd
import matplotlib.pyplot as plt

from sklearn import datasets
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn import preprocessing

from sklearn import tree

# DECISION TREE REGRASSION
from sklearn.tree import DecisionTreeRegressor

#BAGGING AND BOOSTING 
from sklearn.ensemble import BaggingClassifier # Bagging combines the results of multiple models to get a generalized result.
from sklearn.ensemble import AdaBoostClassifier # Boosting method attempts to correct the errors of previous models.

# MODEL BUILDING 
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split

from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import accuracy_score
# WARNING 
import warnings
warnings.filterwarnings('ignore')
```

```python
# IMPORT THE DATASET 
data =pd.read_csv('Company_Data (2).csv')
data 
```

```python
data['ShelveLoc']= data['ShelveLoc'].astype('category')
data['Urban']= data['Urban'].astype('category')
data['US']= data['US'].astype('category')
```

```python
data.dtypes
```

```python
sales_mean = data.Sales.mean()
sales_mean
```

```python
data['High']= data.Sales.map(lambda x: 1 if x > 8 else 0)  
data['High']
```

```python
# label encoder means to convert categorical data into numerical column
label_encoder = preprocessing.LabelEncoder()  # label encoder means to convert categorical data into numerical column
label_encoder
```

```python
data['ShelveLoc'] =label_encoder.fit_transform(data['ShelveLoc'])
data['Urban'] = label_encoder.fit_transform(data['Urban'])
data['US'] = label_encoder.fit_transform(data['US'])
```

```python
data.head()
```

```python
x = data.iloc[:,1:11]
x.head()
```

```python
y = data['High']
y.head()
```

```python
# building the model to train  and test the data 
```

```python
# spliting data into training and testing data set 

x_train, x_test,y_train,y_test = train_test_split(x,y, test_size=0.2,random_state=40)
```

# 1. Building Decision tree classifier using Entropy criteria 

```python
model_1 = DecisionTreeClassifier(criterion = 'entropy',max_depth=3)
model_1.fit(x_train,y_train)
```

```python
# plot the decision tree 

tree.plot_tree(model_1);
```

```python
fn = ['CompPrice','Income','Advertising','Population','Price','ShelveLoc','Age','Education','Urban','US']  # we will extract the feature
cn = ['Sales is high', 'Sales is low']
fig,axes = plt.subplots(nrows=1, ncols=1, figsize= (4,4), dpi= 500)
tree.plot_tree(model_1,
             feature_names=fn,
             class_names= cn,
              filled = True);
```

```python
#Predicting on test data
Preds_1 = model_1.predict(x_test) # predicting on test data set 
Preds_1
```

```python
pd.crosstab(y_test,preds_1) # getting the 2 way table to understand the correct and wrong predictions
#to understand the correct and wrong prediction 
```

```python
print(classification_report(preds_1, y_test))
```

```python
# Accuracy 
np.mean(preds_1==y_test)*100
```

```python
print("Accuracy using C5.0 is: ",accuracy_score(preds_1,y_test)*100)
```

# 2 Building decision tree classifier using gini criteria 

```python
model_gini = DecisionTreeClassifier(criterion='gini', max_depth=3)
model_gini.fit(x_train, y_train)
```

```python
#plot the Decision tree
tree.plot_tree(model_gini);
```

```python
fn = ['CompPrice','Income','Advertising','Population','Price','ShelveLoc','Age','Education','Urban','US']  # we will extract the feature
cn = ['Sales is high', 'Sales is low']
fig,axes = plt.subplots(nrows=1,ncols= 1, figsize = (4,4), dpi= 400)
tree.plot_tree(model_gini,
               feature_names= fn,
               class_names=cn,
               filled = True);
```

```python
#predicting on data set
y_preds_2 = model_gini.predict(x_test)
y_preds_2
```

```python
pd.Series(y_preds_2).value_counts()
```

```python
np.mean(y_preds_2==y_test)
```

```python
print("Accuracy CART: " ,accuracy_score (y_test,y_preds_2)*100)
```

# model by -Bagging 

```python
Dc_Model = BaggingClassifier(DecisionTreeClassifier(max_depth=4), random_state = 0) # Decision Tree classifier object (Bagging)
Dc_Model.fit(x_train, y_train)
```

```python
y_preds_3 = Dc_Model.predict(x_test)
y_preds_3
```

```python
np.mean(y_preds_3==y_test)*100
```

```python
print("Accuracy Bagging Classifier: ", accuracy_score(y_test,y_preds_3)*100 )
```

```python
print(confusion_matrix(y_test,y_preds_3))
print(classification_report(y_test,y_preds_3))
```

# model by - Boosting 

```python
dcmodel_4 = AdaBoostClassifier(DecisionTreeClassifier(max_depth=4), random_state=0) # Decision tree classifier object
dcmodel_4.fit(x_train,y_train)
```

```python
y_preds_4 = dcmodel_4.predict(x_test)
y_preds_4
```

```python
np.mean(y_preds_4==y_test)
```

```python
print("Accuracy by AdaBoost Classifier: ", accuracy_score(y_preds_4,y_test)*100)
```

```python
print(confusion_matrix(y_test, y_preds_4))
print(classification_report(y_test,y_preds_4))
```
