```python
import pandas as pd
import numpy as np
```

```python
book_r= pd.read_csv("book.csv",encoding = "ISO-8859-1")
book_r
df= pd.DataFrame(book_r)
book_r.head()
```

```python
# cleanning the unwanted file
clean=df.drop(['Unnamed: 0'],axis=1)
clean
clean.head()
```

```python
# removing dot in the column 
newname = clean.rename(columns = {'Unnamed: 0':'unnamed','User.ID':'userid','Book.Title':'booktitle','Book.Rating':'bookrating'})
newname
```

```python
len(newname.userid.unique())
```

```python
len(newname.booktitle.unique())
```

```python
# arranging the data
user_newname = newname.pivot_table(index ="userid",
                                   columns ="booktitle",
                                   values ="bookrating").reset_index(drop=True)
user_newname 
```

```python
user_newname.index = newname.userid.unique()
user_newname
```

```python
user_newname.fillna(0,inplace=True)
user_newname
```

```python
# calculate the cosine similarity between the user

from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_similarity = 1 - pairwise_distances(user_newname.values,metric='cosine')
user_similarity
```

```python
# store the result in the data frame
user_similarity_df = pd.DataFrame(user_similarity)
user_similarity_df
```
