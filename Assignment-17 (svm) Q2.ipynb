```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.decomposition import PCA
```

```python
salarydata = pd.read_csv("SalaryData_Train(1) (1).csv")
salarydata
```

```python
salarydata.columns
```

```python
salarydata[salarydata.duplicated()]
```

```python
salarydata1 = salarydata[~salarydata.duplicated()]
salarydata1
```

```python
salarydata1.info()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
salarydata1["Salary"]=le.fit_transform(salarydata1["Salary"])
salarydata1.head()
```

```python
salarydata2 = salarydata1.copy()
salarydata2.head()
```

```python
salarydata1['workclass']=le.fit_transform(salarydata1['workclass'])
salarydata1['education']=le.fit_transform(salarydata1['education'])
salarydata1['maritalstatus']=le.fit_transform(salarydata1['maritalstatus'])
salarydata1['occupation']=le.fit_transform(salarydata1['occupation'])
salarydata1['relationship']=le.fit_transform(salarydata1['relationship'])
salarydata1['race']=le.fit_transform(salarydata1['race'])
salarydata1['sex']=le.fit_transform(salarydata1['sex'])
salarydata1['native']=le.fit_transform(salarydata1['native'])
```

```python
salarydata1.head()
```

```python
x=salarydata1.iloc[:,:13]
y=salarydata1.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc =StandardScaler()
x_train_std = sc.fit_transform(x_train)
x_test_std = sc.fit_transform(x_test)
```

```python
x_train_std
```

```python
x_test_std
```

# SVM classification

```python
svm = SVC (kernel = "rbf",random_state = 0,gamma=0.10,C=1)
```

```python
svm.fit(x_train_std,y_train)
```

```python
y_pred = svm.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
acc = accuracy_score(y_test,y_pred)*100
print("Accuracy=",acc)
confusion_matrix(y_test,y_pred)
```

# changing some values of gamma and c,and cheak results

```python
svm1 = SVC (kernel ="rbf",random_state =0,gamma="auto",C = 10)
svm1.fit(x_train_std,y_train)
```

```python
y_pred_1 = svm1.predict(x_test)
y_pred_1
```

```python
acc_1 = accuracy_score(y_test,y_pred_1)*100
print("Accuracy=",acc_1)
confusion_matrix(y_test,y_pred_1)
```

```python
svm2 = SVC (kernel ="sigmoid",random_state =0,gamma="auto",C = 10)
svm2.fit(x_train_std,y_train)
```

```python
y_pred_2 = svm2.predict(x_test)
y_pred_2
```

```python
acc_2 = accuracy_score(y_test,y_pred_2)*100
print("Accuracy=",acc_2)
confusion_matrix(y_test,y_pred_2)
```
